```python
from scapy.all import sniff
from scapy.layers.inet import IP, TCP, UDP, ICMP

# Function that processes each captured packet
def packet_callback(packet):
    # Check if the packet contains an IP layer
    if IP in packet:
        ip_layer = packet[IP]
        src_ip = ip_layer.src
        dst_ip = ip_layer.dst
        protocol = ip_layer.proto

        # Determine if the packet is using TCP, UDP, or ICMP
        if TCP in packet:
            proto = "TCP"
        elif UDP in packet:
            proto = "UDP"
        elif ICMP in packet:
            proto = "ICMP"
        else:
            proto = f"Other (Protocol {protocol})"

        # Print the source and destination IP addresses and protocol
        print(f"Source IP: {src_ip} -> Destination IP: {dst_ip}")
        print(f"Protocol: {proto}")

        # Check for payload in TCP packets
        if packet.haslayer(TCP):
            tcp_payload = bytes(packet[TCP].payload)
            if tcp_payload:
                print(f"TCP Payload: {tcp_payload}\n")
            else:
                print("No TCP Payload\n")

        # Check for payload in UDP packets
        if packet.haslayer(UDP):
            udp_payload = bytes(packet[UDP].payload)
            if udp_payload:
                print(f"UDP Payload: {udp_payload}\n")
            else:
                print("No UDP Payload\n")

if __name__ == "__main__":
    print("Starting packet capture...")
    # Capture packets and send each one to packet_callback function
    sniff(prn=packet_callback, store=0)
```
